```python
from chemdataextractor.scrape import Selector
from chemdataextractor.scrape.pub.rsc import RscHtmlDocument

htmlstring = open("Ageing mechanisms in electrochemical capacitors with aqueous redox-active electrolytes - ScienceDirect.html").read()
sel = Selector.from_text(htmlstring)
scrape = RscHtmlDocument(sel)
print(scrape.publisher)
scrape.serialize()
```

Here we have been able to get publisher from a completely unknown page
